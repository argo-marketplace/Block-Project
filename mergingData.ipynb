```python
import pandas as pd
import geopandas as gp
```

```python
#Household and families
df = pd.read_csv('aff_download/ACS_16_5YR_S1101_with_ann.csv', header=1)
for idx, name in enumerate(df.columns):
    print idx, name
```

```python
household = df[[ 'Id2', 'Geography', 'Total; Estimate; Total households', 'Total; Margin of Error; Total households',
               'Total; Estimate; HOUSING TENURE - Owner-occupied housing units', 'Total; Margin of Error; HOUSING TENURE - Owner-occupied housing units',
               'Total; Estimate; HOUSING TENURE - Renter-occupied housing units', 'Total; Margin of Error; HOUSING TENURE - Renter-occupied housing units']]

household = household[household.Geography.str.contains('King')]
household.drop(['Geography'], axis=1, inplace=True)
household.reset_index(inplace=True, drop=True)
household.head()
```

```python
#Educational Attainment
df = pd.read_csv('aff_download/ACS_16_5YR_S1501_with_ann.csv', header=1)
for idx, name in enumerate(df.columns):
    print idx, name
```

```python
#Educational Attainment
education = df[['Id2', 'Geography', 'Total; Estimate; Population 25 years and over', 'Total; Margin of Error; Population 25 years and over',
                'Total; Estimate; Population 25 years and over - Less than 9th grade', 'Total; Margin of Error; Population 25 years and over - Less than 9th grade',
                'Total; Estimate; Population 25 years and over - 9th to 12th grade, no diploma', 'Total; Margin of Error; Population 25 years and over - 9th to 12th grade, no diploma',
                'Total; Estimate; Population 25 years and over - High school graduate (includes equivalency)', 'Total; Margin of Error; Population 25 years and over - High school graduate (includes equivalency)',
                'Total; Estimate; Population 25 years and over - Some college, no degree', 'Total; Margin of Error; Population 25 years and over - Some college, no degree',
                'Total; Estimate; Population 25 years and over - Associate\'s degree', 'Total; Margin of Error; Population 25 years and over - Associate\'s degree',
                'Total; Estimate; Population 25 years and over - Bachelor\'s degree', 'Total; Margin of Error; Population 25 years and over - Bachelor\'s degree',
                'Total; Estimate; Population 25 years and over - Graduate or professional degree', 'Total; Margin of Error; Population 25 years and over - Graduate or professional degree']]
education = education[education.Geography.str.contains('King')]
education.drop(['Geography'], axis=1, inplace=True)
education.reset_index(inplace=True, drop=True)
education.head()
```

```python
#Median Household income
df = pd.read_csv('aff_download/ACS_16_5YR_S1903_with_ann.csv', header=1)
for idx, name in enumerate(df.columns):
    print idx, name
```

### Do we want to focus on high income neighbourhoods with high senior population or just check the general population

```python
medhhinc = df[['Id2', 'Geography', 'Median income (dollars); Estimate; Households', 'Median income (dollars); Margin of Error; Households']]
medhhinc = medhhinc[medhhinc.Geography.str.contains('King')]
medhhinc.drop(['Geography'], axis=1, inplace=True)
medhhinc.reset_index(inplace=True,drop=True)
medhhinc.head()
```

```python
finalDf = pd.merge(household, education, on='Id2', how='outer', suffixes=('_household', '_education'))
finalDf = pd.merge(finalDf, medhhinc, on='Id2', how='outer', suffixes=('_hhOrEdu', '_medhhinc'))
finalDf.head()
```

```python
morgatestatus = gp.read_file('morgatgestatusshp/acs2016_5yr_B25081_14000US53033009200/acs2016_5yr_B25081_14000US53033009200.shp')
cols = {u'B25081001' : 'Total: Mortgage Status', u'B25081001e' : 'Total Mortgage Status Error', 
        u'B25081002' : 'Housing units with a mortgage, contract to purchase, or similar debt', u'B25081002e' : 'Housing units with a mortgage, contract to purchase, or similar debt: Error',
        u'B25081003' : 'With either a second mortgage or home equity loan, but not both', u'B25081003e' : 'With either a second mortgage or home equity loan, but not both: Error', 
        u'B25081004' : 'Second mortgage only', u'B25081004e' : 'Second mortgage only: Error',
        u'B25081005' : 'Home equity loan only', u'B25081005e' : 'Home equity loan only: Error', 
        u'B25081006' : 'Both second mortgage and home equity loan', u'B25081006e' : 'Both second mortgage and home equity loan: Error',
        u'B25081007' : 'No second mortgage and no home equity loan', u'B25081007e' : 'No second mortgage and no home equity loan: Error',
        u'B25081008' : 'Housing units without a mortgage', u'B25081008e' : 'Housing units without a mortgage: Error'}
morgatestatus = morgatestatus.rename(columns=cols)
morgatestatus['Id2'] = morgatestatus.geoid.apply(lambda x: x.split('US')[1])
morgatestatus.Id2 = pd.to_numeric(morgatestatus.Id2)
morgatestatus.drop(['name', 'geoid'], axis=1, inplace=True)
morgatestatus.columns
```

```python
finalDf = pd.merge(finalDf, morgatestatus, on='Id2', how='outer',suffixes=('_finalDf', '_morgatestatus'))
finalDf = gp.GeoDataFrame(finalDf)
finalDf.crs = {'init': u'epsg:4326'}
finalDf.columns
```
